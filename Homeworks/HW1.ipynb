<div dir="rtl">
    روش اول حذف ستونی‌ست که تعداد داده‌های گم‌شده در آن ستون زیاد است. برای نمونه در مثال قبل، ستون‌های Competitors, Easy Apply نمونه‌ای از این ستون‌ها هستند. ستون‌هایی با حدود ۷۵ و ۹۰ درصد داده‌ی گم‌شده. به وضوح این ستون‌ها نمی‌توانند اطلاعات مفیدی برای ما فراهم کنند. مورد بعدی حذف سطرهای دارای مقدار گم‌شده است. در تمامی این حذف‌ها باید دقت کرد، تناسبی بین تمامی اطلاعات موجود و اطلاعاتی که حذف می‌شوند وجود داشته باشد. به این معنی که اگر با این حذف‌ها نصف داده‌ها را از بین ببریم شاید این کار مناسبی نباشد و تحلیل ما بر روی این داده‌ها را تحت تاثیر قرار دهد. به طور کلی برای انجام حذف باید به محیط توجه کرد و اینکه در نظر داشت با این حذف‌ها به اطلاعات خدشه‌ای وارد نشود. ایده‌ی بعدی که برای داده‌های عددی مناسب است قرار دادن میانگین آن ستون به جای داده‌ی گم‌شده می‌باشد. بدین شکل که اگر مقدار سن فردی در یک ستون فرضی ناموجود بود، این سن را با میانگین سن افراد دیگر جایگزین کرده تا کمترین تاثیر را در پارامترهای ارزیابی داده‌ها مانند میانگین و واریانس داشته باشیم. مورد دیگری نیز که می‌توان از آن بهره برد، استفاده از الگوریتم نزدیک‌ترین همسایه می‌باشد. بدین ترتیب که برای مثال اگر در سطری از داده‌ها مقدار حقوق شغلی ناموجود بود. با بررسی سطرهای دیگر به پیدا کردن نزدیک‌ترین و شبیه‌ترین سطر به سطر حاوی داده گم‌شده می‌پردازیم و مقدار آن متغییر را برابر متغییر موجود در سطر نزدیک به آن می‌گذاریم. این کار تا حدودی با شهود ما نیز سازگار است. این که اگر مقادیر متناظر این سطر با سطری نزدیک بود، بنابراین حقوق آن‌ها نیز باید مقداری از همان مرتبه باشد. همچنین ایده‌ی دیگری که برای کار با این داده‌های گم‌شده وجود دارد به منظور استفاده از این داده‌ها در الگوریتم‌های یادگیری ماشین این است که از الگوریتم‌هایی استفاده کنیم که مشکلی با داشتن داده‌هایی با مقدار گمشده ندارند. مانند k نزدیک‌ترین همسایه یا naive Base. 
    
    
</div>

<div dir="rtl">

در متدهای ذکر شده باید در نظر داشت تا اولاً داده‌ها آسیب نبینند. مانند حذف درصد قابل توجهی از داده‌ها یا تغییر معناداری در متغییرهای عددی. ثانیا سمنتیک محیط تعیین‌کننده‌ی معنی‌دار بودن یا نبودن رویکرد ما نسبت به هر متغییر است. برای مثال گرفتن میانگین یا الگوریتم نزدیک‌ترین همسایه بر روی شماره شناسنامه کار بی‌معنایی‌ست، در حالی که همین کار برای سن یا دستمزد اتفاق دور از ذهنی نیست. 

</div>

<div dir="rtl">
تا اینجای کار فقط داشتیم داده‌هامون رو تمیز می‌کردیم. از اینجا به بعد می‌خوایم کمی روی اونها کار کنیم. یکی از 
فاکتورهای مهم توی کار کردن میزان حقوقه. پس بیاید و ستون تخمین حقوق رو به دو ستون تبدیل کنید که حداقل و حداکثر حقوق 
درخواستی در اون درج شده.
سپس ستون میانگین حقوق را از این دو ستون جدید به دست بیارید و در انتها پنج داده‌ی اول مجموعه داده رو نمایش بدید.
    (۸ امتیاز)
</div>

```python
df['avg_salary'] = df['Salary Estimate'].apply(lambda x: re.search('\$(.*)K', x).group(1))
df['min_salary'] = df['avg_salary'].apply(lambda x: int(re.search('(.*)K', x).group(1)))
df['max_salary'] = df['avg_salary'].apply(lambda x: int(re.search('\$(.*)', x).group(1)))
df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2
df.drop('Salary Estimate', axis=1, inplace=True)

df.head(5)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.shape
```

<div dir="rtl">
    می‌خوایم یه آمار کلی از حقوق کارها داشته باشیم. کاری که می‌خوایم بکنید اینه که ۵٪ از کارها با بالاترین میانگین حقوق و ۵٪ از کارها با پایین‌ترین میانگین حقوق رو حذف کنید و بعدش یه آمار کلی از میانگین حقوق‌ها بدید و همچنین بگید که بنظرتون چرا ۵٪ از بالا و ۵٪ از پایین رو حذف کردیم؟ آمار کلی‌ای که می‌خوایم شامل:  mean, std, min, max می‌شه.
    (۶ امتیاز)
</div>

<div dir="rtl">
حقوق‌های بسیار بالا و بسیار پایین عموما منحصر به افراد معدودی می‌شود. برای مثال مدیران عالی‌رتبه که تعدادشان در برابر تعداد کارمندها انگشت‌شمار است و به همین شکل کارمندان تازه وارد. بدین ترتیب ممکن است فاصله‌ی معنی‌داری میان حقوق این افراد از جمعیت عمومی وجود داشته باشد و با این کار تاثیر زیادی بر پارامترهای ارزیابی ستون عددی دستمزد بگذارند. به همین علت با حذف این داده‌های به اصطلاح پرت از داده‌های اصلی به دریافت بهتر اطلاعات توسط پارامترهای ارزیابی کمک می‌کنیم. همچنین گاهی ممکن است به اشتباه برخی اعداد وارد شوند که در حالت درصد خطای اندک، اگر این داده‌ها در میانه‌ی داده‌ها باشند تاثیر چندانی بر پارامترهای ارزیابی نظیر میانگین و میانه و ... نمی‌گذارند. اما اگر همین داده‌های خطا بسیار از داده‌های اصلی دور باشند می‌توانند تاثیر چشم‌گیری بر این پارامترها بگذارند. برای مثال نمره افراد یک کلاس را در نظر بگیرید که عموما در بازه‌ی ۱۴ تا ۲۰ است. حال فرض کنید به اشتباه نمره‌ی فردی به جای ۱۹ عدد ۱۱۹ وارد شود. تاثیر همین یک نمره بر میانگین یک کلاس را می‌توان حدس زد. به همین شکل برای داده‌های دیگر :) 
</div>

```python
df.sort_values(by=['avg_salary'], inplace=True, ignore_index=True)
df = df[df.avg_salary < df.avg_salary.quantile(0.95)]
df = df[df.avg_salary > df.avg_salary.quantile(0.05)]
df.reset_index(drop=True, inplace=True)
print(df.shape)
print(df.describe())
df
```
